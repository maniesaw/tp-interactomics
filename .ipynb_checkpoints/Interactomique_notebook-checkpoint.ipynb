pubid:(17446270)

http://www.ebi.ac.uk/Tools/webservices/psicquic/intact/webservices/current/search/query/pubid:(17446270)?format=tab27



```python
import requests
url = "http://www.ebi.ac.uk/Tools/webservices/psicquic/intact/webservices/current/search/query/pubid:(17446270)?format=tab27"

try:
    httpReq = requests.get(url, proxies=None)
except NameError:
    httpReq = requests.get(url)
ans = httpReq.text
```

```python
print(ans)
```

### Filtre sur les lignes avec identifiant 

```python
Ans_split = ans.split("\t")
Ans_split
```

```python
Ans_Matrix = [ i.split("\t") for i in ans.split("\n") if len(i.split("\t"))>0]
Ans_Matrix
```

```python
Ans_Uniprot = [ i for i in Ans_Matrix if not i[0]=="" and not i[1]=="" ]
Ans_Uniprot
len(Ans_Uniprot)
```

```python
# Recuperation des methodes
Set_method = [i[6] for i in Ans_Uniprot]
set(Set_method)
```

```python
Ans_EBV = [ i for i in Ans_Uniprot if (i[9].split("(")[1].split(")")[0][0:3] =="ebv" and i[10].split("(")[1].split(")")[0][0:3]=="ebv") ]
Ans_EBV
len(Ans_EBV)
```

```python
Ans_EBV_Human_And = [ i for i in Ans_Uniprot if (i[9].split("(")[1].split(")")[0][0:3] =="hum" and i[10].split("(")[1].split(")")[0][0:3]=="hum") ]
Ans_EBV_Human_And
len(Ans_EBV_Human_And)
```

```python
Ans_EBV_Human = [ i for i in Ans_Uniprot if (i[9].split("(")[1].split(")")[0][0:3] =="hum" or i[10].split("(")[1].split(")")[0][0:3]=="hum") ]
Ans_EBV_Human
len(Ans_EBV_Human)
```
